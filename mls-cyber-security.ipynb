## Abstract
***

- applying the methodologies of machine learning (``ml``) to cybersecurity problems [1]

- ``ml`` is within an area of Artificial intelligence (``ai``) [1]

- within ``ml`` is Deep Learning (``dl``) which is a subset of ``ml`` [1]

- datasets were of a certain size in the beginning so that ``ml`` could handle them [1]

- it started to get so big that new techniques had to be developed therefore the ``ml`` algorithms (in particular ``dl`` algorithms) required development of tools that could handle really large amounts of data (big data) [1]

- big data required the ``ml`` algorithms (``dl`` algorithms) to evolve to become something else that can process that data [1]

- ``ml`` can be applied to several areas such as health care and traditional information retrieval (google is indexing all the worlds websites therefore indexing a lot of text) [1]

- nowadays these have been applied to a lot of tools such as alexa (can be spoken to) and self driving cars (autonomous vehicles) and so many application areas (there is no limit) - one of those areas is of course cybersecurity [1]

- people may ask the question so what is the algorithm in ``ml`` for cybersecurity and the answer is that there is no specific algorithm for cybersecurity [1]

- ``ml`` or data science is just a tool set of algorithms and those algorithms can be applied equally to healthcare or alexa or google or self driving cars or cybersecurity [1]

- the difference is the data - the data is going to be from sources related to cybersecurity issues such as network packets and/or log files and/or the files themselves etc [1]

<a id="introductionVectorSpaceModel"></a>
***
### Vector Space Model
***

- the data in healthcare might be images of cancer - the data in alexa is text - the data in self driving cars are sensors [1]

- all those different kinds of information which are in different formats and in different sources and different origins need to be converted into one specific format and that format is called the vector space model (``vsm``) [1]

- all can be converted into the ``vsm`` - this is the one common theme they all have regardless of the dataset [1]

- once something is converted into the ``vsm`` then it can be applied to ``ml`` [1]

- ``ml`` algorithms should be understood but do not really need to understand them because its a matter of just using the tools to use the models especially once an algorithm is employed the same code is the same [1]

- whenever faced with a new dataset will always have to go from whatever it is to the ``vsm`` and that is the challenge [1]

## Introduction
***

- ``ai`` is a more broader term and does not necessarily always learn from data (can have heuristic rules or other ways of learning) whereas ``ml`` is more focused on learning on data [1]

- ``ml`` are methods for predicting and/or detecting and/or grouping data samples based on a model - the model must be learned with data (from samples with associated labels to learn correlations) [1] 

- as for the models the methods can be geometrical (or not) [1]

- referring to geometrical models these can be based on distance metrics and linear boundaries [1]

- could say that algorithms are algebra and ``ml`` is geometry in a way albeit usually when referring to the idea of the ``vsm`` think about points and shapes and building lines or curves in those spaces therefore there is the concept of geometry involved [1]

< [Table of Contents](#top) | [References](#references) >
<a id="introductionCybersecurity"></a>
***
### Cybersecurity
***

- ``ml`` is the idea of learning from data [1]

- cybersecurity usually requires expert knowledge in a specific area (specific type of an attack) but that requires a person to analyse the data for a long time [1]

- nowadays there is so much data that automatisation solutions are required [1]

- need some way of having computers help with all the hard work - building models by hand is labour intensive so ``ml`` could learn the models to find patterns in the data etc (is more efficient) [1] 

- a laptop might be good for prototyping a small dataset but with big data other solutions need to be employed like GPUs and dedicated computers [1]

- need more powerful computers or better yet several computers working in parallel (one of the ideas of big data) [1]

- there are two main approaches: i. parallel CPUs with eight cores (hadoop - idea of creating clusters) or ii. GPUs whereby one GPU (maybe three or four) on the same motherboard can have thousands of cores [1]

- both are different processors - the CPU can do more things whereas a GPU can only do one specific thing - it so happens that the specific thing is one of the key components needed to do ``ml`` which is matrix multiplication [1]

<a id="introductionSupervisedLearning"></a>
***
###### Supervised Learning - Classifiers
***

- have: i. samples (network traffic data ) and ii. labels (good or bad) [1]

- the basic pipeline requires taking the ``Data`` and dividing it into: i. ``Train Data`` and ii. ``Test Data`` [1]

- building the model from the ``Train Data`` by selecting a ``Classifier`` [1]

- once the ``Train Model`` is built then going to take the test dataset (``X_test``) and apply it to the ``Train Model`` [1]

- the test dataset (``X_test``) and its associated labels will be evaluated (``y_test`` labels) [1]

- the model takes the ``X_test`` and runs it through the ``Train Model`` performing prediction (``y_pred``) - in the evaluation comparing both (same or not) [1]

<a id="introductionUnsupervisedLearningClustering"></a>
***
###### Unsupervised Learning - Clustering
***

- have: i. samples (network traffic data ) and no labels [1]

- cannot learn correlations but can learn to group things [1]

- with the ``vsm`` might argue the existence of two clusters - cannot really learn a model to predict things necessarily but can at least group the different points (technique associated with clustering) [1]

### Machine Learning Pipeline
***

- pre-processing the data (format/clean etc) [1]

- build the ``vsm`` (represents the data geometrically) [1]

- a lot of algorithms require the dataset to be in a certain format and once formatted then can easily fit it into all solutions [1]

- once the data is in the ``vsm`` then can apply any machine learning algorithm (including deep learning) [1]

- using supervised learning will end up with a model that predicts some data - comparing the prediction to the real data therefore measuring its performance in the evaluation stage [1]
